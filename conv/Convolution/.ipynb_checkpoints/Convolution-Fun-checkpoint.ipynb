```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
import cv2
```

```python
im = cv2.imread("quad.jpg")
# cv2.imshow('image',im)
# cv2.waitKey()
# cv2.destroyAllWindows()
```

```python
print(im.shape)
w,h = im.shape[0:2]
print(w,h)
```

```python
def grayscale(img):
    _root = la.norm(np.asarray([255,255,255]))
    bw_im = np.asarray([[la.norm(img[i,j])/_root for j in range(h)] for i in range(w)])
    return bw_im

bw_im = grayscale(im)
```

```python
# cv2.imshow('image',bw_im)
# cv2.waitKey()
# cv2.destroyAllWindows()

plt.figure(figsize=(12,9))
plt.imshow(bw_im)
plt.show()
```
