```python
n = 256
k = 4
Q = np.zeros((2*n-1,(2*n//k-1)*(2*k-1)))
s = 2*(n//k)-1
for i in range(2*k-1):
    x = (n//k)*i
    y = ((2*n-k)//k) * i
    Q[x:x+s,y:y+s] = np.eye(s)
```

```python
def Toom_k(f,g,k,E,I):
    n = len(f)
    assert(n == len(g))
    p = round(log(n,k))
    assert(k**p == n)
    
    if n == 1: return f*g
    
    # fold into a quasi-tensor
    F = np.reshape(f, newshape=(k,-1))
    G = np.reshape(g, newshape=(k,-1))
    
    # evaluate 
    P = np.dot(E,F)
    Q = np.dot(E,G)
    x,y = P.shape
    
    # multiplication (recursion)
    R = np.zeros((x,2*y-1))
    for i in range(x):
        R[i] = Toom_k(P[i],Q[i],k,E,I)
        
    # interpolate
    T = np.dot(I,R)

    # recompose
    t = np.reshape(T, newshape=(-1,)) 
    recomp = QMat(n,k)
    return np.dot(recomp, t)
```

### Direct v. DFT

```python
def directConv(F,G):
    n = len(G); b = len(F)
    z = n + b - 1
    Y = np.zeros((z,z))
    for x in range(z):
        for y in range(z):
            for i in range(b):
                for j in range(b):
                    Y[x,y] += 0 if x-i<0 or x-i>=n \
                    or y-j<0 or y-j>=n \
                    else F[i,j]*G[x-i,y-j]
    return Y

def convByDFT(F,G):
    n = len(G); b = len(F)
    z = n + b
    F2 = np.zeros((z,z))
    G2 = np.zeros((z,z))
    F2[:b,:b] = F
    G2[:n,:n] = G
    Fr = Fmat(z)
    Frinv = Fmatinv(z)
    
    Y = (Frinv @ ( (Fr @ F2 @ Fr.T) * (Fr @ G2 @ Fr.T) ) @ Frinv.T).real
    return(Y[:-1,:-1])
```

### Monomial with Integer and Chebyshev Nodes, Chebyshev

```python
def interpolantMats(n,b,cheby=False):
    z = n+b-1
    pts = rs(z)
    
    if cheby:
        i = np.arange(1,z+1, dtype=np.float64)
        pts = np.cos((2*i-1)/(2*z)*np.pi)
        
    V = np.vander(pts,increasing=True)
    if not cheby:
        V[-1,:] = 0
        V[-1,-1] = 1
    
    C = la.inv(V)
    A = V[:,:b].copy()
    B = V[:,:n].copy()
    
    if not cheby:
        A[-1,-1] = 1
        B[-1,-1] = 1
        
    return [C,A,B]

def tc1(F,G,cheby=False):
    C,A,B = interpolantMats(len(G),len(F),cheby)
    return C @ ((A @ F @ A.T) * (B @ G @ B.T)) @ C.T

def chebyMat(n,b):
    # n is number nodes, b is number of basis
    i = np.arange(1,n+1, dtype=np.float64)
    nodes = np.cos((2*i-1)/(2*n)*np.pi)
    basis = np.arange(b, dtype=np.float64)
    
    # (basis,nodes)
    return np.cos(basis * np.arccos(nodes.reshape(-1, 1)))

def interpolantCheby(n,b):
    R = n+b
    RR = 2*R-1
    MC = chebyMat(RR,RR)
    C = la.inv(MC)[-(R-1):,:]*2
    A = chebyMat(RR,R)[:,-b:]
    B = chebyMat(RR,R)[:,-n:]
    return [C,A,B]

def tc_cheby(F,G):
    C,A,B = interpolantCheby(len(G),len(F))
    return C @ ((A @ F @ A.T) * (B @ G @ B.T)) @ C.T
```

### Nested Formula

```python
def nestedMats(F,G,factors,cheby=False):
    n = len(F)
    prods = 1
    for i in factors: prods *= i
    assert(prods == n)
    
    k = factors[0]
    c,a,b = interpolantMats(k,k,cheby)
    
    C = np.kron( np.eye(1) , 
                    np.dot(QMat(n, k), 
                        np.kron(c, np.eye(2*n//k-1) ) 
                    ) 
                )
    
    A = a.copy()
    B = b.copy()
    
    preTotal = k
    postTotal = 2*k-1
    
    for k in factors[1:]:
        c,a,b = interpolantMats(k,k,cheby)
        A = np.kron(A,a)
        B = np.kron(B,b)
        
        C = np.dot(C, np.kron( np.eye(postTotal) , 
                        np.dot(QMat(n//preTotal, k), 
                            np.kron(c, np.eye(2*n//(preTotal*k)-1) ) 
                            ) 
                        )
                    )
        preTotal *= k
        postTotal *= 2*k-1
        
    # print(C.shape,A.shape,B.shape)
    return C,A,B

def nestedConv(F,G,cheby=False):
    n = len(F)
    C,A,B = nestedMats(F,G,primeList(n),cheby)
    return C @ ((A @ F @ A.T) * (B @ G @ B.T)) @ C.T

def primeList(x):
    if x == 1: return np.asarray([1])
    for f in range(2,x):
        if x % f == 0:
            return np.append(f,primeList(x//f))
    return np.asarray([x])
```

### 3D

```python
def contract3D(A,M):
    K = np.einsum("ik,kyz->iyz",A,M)
    K = np.einsum("ik,xkz->xiz",A,K)
    K = np.einsum("ik,xyk->xyi",A,K)
    return K

def tc1_3D(F,G,cheby=False):
    C,A,B = interpolantMats(len(G),len(F),cheby)
    F2 = contract3D(A,F)
    G2 = contract3D(B,G)
    Y = contract3D(C,F2*G2)
    return Y

def tc_cheby_3D(F,G):
    C,A,B = interpolantCheby(len(G),len(F))
    F2 = contract3D(A,F)
    G2 = contract3D(B,G)
    Y = contract3D(C,F2*G2)
    return Y

def nestedConv_3D(F,G,cheby=False):
    n = len(F)
    C,A,B = nestedMats(F,G,primeList(n),cheby)
    F2 = contract3D(A,F)
    G2 = contract3D(B,G)
    Y = contract3D(C,F2*G2)
    return Y

def winoConv_3D(polys,F,G):
    n,r = len(G),len(F)
    [C,A,B] = winoMats(polys,n,r)
    [C,A,B] = moduloToLinear(polys,n,r,C,A,B)
    F2 = contract3D(A,F)
    G2 = contract3D(B,G)
    Y = contract3D(C,F2*G2)
    return Y

def directConv_3D(F,G):
    n = len(G); r = len(F)
    R = n + r - 1
    Y = np.zeros((R,R,R))
    for x in range(R):
        for y in range(R):
            for z in range(R):
                for i in range(r):
                    for j in range(r):
                        for h in range(r):
                            Y[x,y,z] += 0 if x-i<0 or x-i>=n \
                            or y-j<0 or y-j>=n \
                            or z-h<0 or z-h>=n \
                            else F[i,j,h]*G[x-i,y-j,z-h]
    return Y

```
