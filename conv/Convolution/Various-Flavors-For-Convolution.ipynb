### Introduction

In this notebook, we will examine the speed and stability of several convolution techniques. In particular, convolution typically works through interpolation. We evaluate at $2n-1$ points, then use interpolation to solve for the coefficients, which is the convolution.

The techniques we've seen and will analyze are (still growing):

1. DFT
1. FFT
1. Karatsuba
1. Toeplitz
1. Toom-k for k > 2
1. Cooley-Tukey FFT with radix > 2

```python
def errorPlot(name, method, mx=100, update=True,twoPower=False,lowOrder=False):
    sizes = np.zeros(0)
    err = np.zeros(0)

    milemark = 0
    for i in range(3,mx):
        if (update and i >= 100*milemark):
            print(i)
            milemark += 1
        if(twoPower):
            if(2**round(log(i,2)) == i):
                sizes = np.append(sizes, i)
                err = _error(i,method,err,lowOrder)
        else:
            sizes = np.append(sizes, i)
            err = _error(i,method,err,lowOrder)

    plt.figure(figsize=(12,8))
    plt.plot(sizes,err)
    plt.xlabel("size")
    plt.ylabel("Error")
    plt.title(name + " Stability")
    
def _error(n,method,err,lowOrder):
    f = randomvec(n,seedNumber=n)
    g = randomvec(n,seedNumber=2*n)

    conv = method(f,g)
    convLib = np.convolve(f,g)

    if(lowOrder):
        return np.append(err, la.norm(conv[:n]-convLib[:n])/la.norm(convLib[:n]))
    return np.append(err, la.norm(conv-convLib)/la.norm(convLib))
```

### DFT

```python
def _dft(f,g):
    n = len(f)
    assert(n == len(g))
    
    F = Fmat(2*n-1)
    Finv = Fmatinv(2*n-1)

    # need to append 0s to the end
    f2 = np.append(f,np.zeros(n-1))
    g2 = np.append(g,np.zeros(n-1))
    
    return (Finv @ ((F @ f2) * (F @ g2))).real

errorPlot("DFT", _dft,100)
```

### FFT

```python
def _fft(f,g):
    n = len(f)
    
    f2 = np.append(f,np.zeros(n))
    g2 = np.append(g,np.zeros(n))
    # return np.fft.ifft(np.fft.fft(f2)*np.fft.fft(g2)).real[:-1]
    # powers of two
    return invfft(fft(f2) * fft(g2)).real[:-1]

powerOfTwos = True
errorPlot("FFT", _fft,1<<11+1,False,powerOfTwos)
```

### FFT Cost

As analyzed earlier, let $(\cdot, \cdot)$ be the cost of addition and multiplication of complex numbers. The total cost is

$$ \big(6nlog(n) + 6n, 3nlog(n) + 5n\big)$$

Note these are complex flops. If $n < b$, then the cost is approximately

$$ \big(6nlog(b) + 6n + 2blog(b) + 2b,3nlogb + 7n + blogb -b) $$

### Karatsuba (Toom-2)

When the size of the filter $f$ and data $g$ are the same, we will turn to Karatsuba to run our algorithm in $O(n^{1.58})$ time.

If we use the implicit form, we can save some computation using FIREHAB.

Note for now, the implemenetation of Karatsuba only handles low order, or the first $n$ terms.

```python
def _karastuba(f,g):
    return FIREHAB(f,g[::-1])

errorPlot("Karatsuba", _karastuba, 1<<11+1,update=False,twoPower=True,lowOrder=True)
```

### Cost

The flops derived earlier is:

$$\le \frac{n^{log_k(4-1/k)}}{3} \cdot \Big( 2E(n,k) + I(n,k) + R(n,k) \Big) = \frac{n^{log_k(4-1/k)}}{3} \cdot (16nk - 6n, 16nk - 8n) $$

$E(n), I(n)$, and $R(n)$ correspond to evaluating, interolating, and recomposing. The first two can be done with matrix multiplication. The third can be done with matrix multiplication, but it might be overkill. The final flop count on the right is assuming naive matrix multiplication.
